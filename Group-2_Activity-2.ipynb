```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Downloads/diabetes_model.csv')
df.head(10)
```

```python
df.isnull().sum()
```

# 1. Create a Count plot of Outcome

```python
plt.figure(figsize=(10,10))
sns.countplot(x = 'Outcome',data = df)
sns.set_theme(style="darkgrid")
plt.show()
```

# 2. Box plot age vs outcome

```python
plt.figure(figsize=(10,10))
sns.boxplot(x = 'Outcome', y= 'Age',data = df)
plt.title("Age vs Outcome")
plt.show()
```

Analysys: The visualization shows a higher chance of Diuabetics with increase in age.

# 3. Check for the chances of getting diabetes if it is common in family, using the Daiabetes Pedigree Function

```python
plt.figure(figsize=(10,10))
sns.boxplot(x='Outcome', y='DiabetesPedigreeFunction',data=df)
plt.title("DPF vs Outcome")
plt.show()
```

Analysy: It is visible that, Chances of getting Diabetics is more likely in families.

# 4. Check for the relationship with Insulin & Glucose and group the datapoints by Outcome.

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x ='Insulin', y='Glucose', hue='Outcome', data=df)
plt.title('Relationship between Insulin and Glucose')
plt.show()
```

Analysys: The visualization shows an almost linear relationship between insulin and Glucose

# 5. Create a heatmap on Correlation of features in the data.

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),annot = True)
plt.xticks(rotation=90, fontsize=16)
plt.yticks(fontsize=16)
plt.show()
```
