```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Downloads/Telco-Customer-Churn.xlsx - Telco-Customer-Churn.csv')
```

```python
df.info()
```

```python
df2 = df[df.TotalCharges!=' ']
```

```python
df2.TotalCharges = pd.to_numeric(df2.TotalCharges)
```

```python
df2
```

# 1) Compare churn count with respect to gender. 

```python
plt.figure(figsize=(10,10))
sns.countplot(x='Churn', hue='gender', data=df)
plt.title("Churn count VS gender", fontsize=16)
plt.show()
```

Churn count seems to be evenly distributed amongst both genders. (i.e, Customer churning behaviour is irrespective of gender)

# 2) Find out how many female senior citizens there in the dataset

```python
df1 = df2[df2['SeniorCitizen']==1]
```

```python
plt.figure(figsize=(10,10))
ax = sns.countplot(x='gender', data=df1)
plt.title("No. of Senior citizens: Male vs Female", fontsize=16)
ax.set(ylim=[500,600])
for p in ax.patches:
    ax.annotate(f'\n{p.get_height()}', (p.get_x()+0.4, p.get_height()), ha='center', va='top', color='black', size=18)
plt.show()    
```

The visualisation tells us that, total female senior citizens in the dataframe are 568

# 3) Compare 'tenure' with 'Total Charges' 

```python
plt.figure(figsize=(20,10))
sns.scatterplot(data=df2, x="tenure", y="TotalCharges")
plt.title("Scatter-Plot on Tenure and Total charges", fontsize=25)
plt.show()
```

From the visualisation, it is clear that, an increase in tenure, increases Total Charges.

# 4) Find out which contract preffered by the senior citizen

```python
a = df2[df2["SeniorCitizen"]==1]
plt.figure(figsize=(10,10))
plt.title("Pie chart on distribution of contract preffered among Senior Citizens ", fontsize=25)
a.Contract.value_counts(sort=True).plot.pie(fontsize = 16, autopct = '%1.1f', explode = [0.1,0,0], startangle = 90)
plt.yticks(rotation =  90, fontsize=16)
plt.show()
```

From the visualization, it is clear that, majority of senior citizens predominantly prefer Month-to-Month contracts.

# 5) Comment your finds on Payment Method?

```python
plt.figure(figsize=(20,10))
ax = sns.countplot(data=df2, x='PaymentMethod')
plt.title("Count Plot of Different Methods of Payment", fontsize=20)
for p in ax.patches:
        ax.annotate('{:.0f}'.format(p.get_height()), (p.get_x()+0.1, p.get_height()+50))
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.histplot(data=df2, x="PaymentMethod", hue="Contract", multiple="dodge", shrink=.8)
plt.title("Count Plot of Different Methods of Payment with regards to Contract", fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.histplot(data=df2, x="PaymentMethod", hue="SeniorCitizen", multiple="dodge", shrink=.8)
plt.title("Count Plot of Different Methods of Payment based on Senior Citizenship", fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.histplot(data=df2, x="PaymentMethod", hue="gender", multiple="dodge", shrink=.8)
plt.title("Count Plot of Different Methods of Payment based on Gender", fontsize=20)
plt.show()
```

Findings from these visualizations are:

1. Electronic checks are widely preffered among the customer base.
2. Longer contract period customers seems to overwhelmingly prefer Credit cards.
3. A large part of the senior citizens have moved to electronic payment methods like E-Checks, Bank transfers and Credit cards.
4. Mode of payment seems to be evenly divided amongst genders. (i.e males or females doesn't seem to have a preffered method of payment method)
